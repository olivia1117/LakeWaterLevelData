```python

import pandas as pd

df = pd.read_excel(r'./Water_Level_Data_Reformatted.xlsx')
print (df)

```

```python
histogram_dataframe = df.loc[df['Lake'] == "Michigan-Huron"]

```

```python
#print(histogram_dataframe)
mean_dataframe = histogram_dataframe.loc[histogram_dataframe["Type"] == "mean"]
print (mean_dataframe)
nineteen_dataframe = histogram_dataframe.loc[histogram_dataframe["Type"] == "2019_raw"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

mean_dataframe['Water Level (Feet)'].hist()

plt.title("Mean Water Level in Michigan-Huron")
plt.xlabel("Water Level in Feet")
plt.ylabel("Number of Months")
```

```python
f, axes = plt.subplots(2, 1, figsize=(7, 7), sharex = True, sharey = True)
sin_annual = mean_dataframe.loc[mean_dataframe["Month"] != "Annual"]
sin_annual.plot.scatter(x ='Month', y ='Water Level (Feet)', title = 'Mean Water Level in Michigan-Huron (by month)', ax = axes [0] )


last_year_data = nineteen_dataframe.loc[nineteen_dataframe["Month"] != "Annual"]
last_year_data.plot.scatter(x ='Month', y ='Water Level (Feet)', title = 'Water Level in Michigan-Huron for 2019 ONLY (by month)', ax = axes[1] )


```

```python
f, axes = plt.subplots(2, 1, figsize=(7, 7), sharex = True, sharey = True)
sin_annual = mean_dataframe.loc[mean_dataframe["Month"] != "Annual"]
sin_annual.plot.line(x ='Month', y ='Water Level (Feet)', title = 'Mean Water Level in Michigan-Huron (by month)', ax = axes [0] )


last_year_data = nineteen_dataframe.loc[nineteen_dataframe["Month"] != "Annual"]
last_year_data.plot.line(x ='Month', y ='Water Level (Feet)', title = 'Water Level in Michigan-Huron for 2019 ONLY (by month)', ax = axes[1] )
```
